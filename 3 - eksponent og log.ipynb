# 3 - Eksponentialfunksjonen, logaritmer og rentegregning

### Importerer nødvendige pakker

```python
import numpy as np
from matplotlib import pyplot as plt
```

### Vi lager en innskuddsfunksjon

Om banken din legger til renten én gang i året og renten er r, så er bankinnskuddet på x=100 om T år gitt som:

```python
def account_balance(x,r,T):
    return x*(1+r)**T
```

Om renten er 5%=0.05, innskuddet og innskuddet er 100, så kan vi tegne innskuddsfunksjonen fra år 10 som

```python
#Vi sjekker utviklingen over 20 år, med 100 datapunkter
T = np.linspace(0,10,100)

#plotter funksjonen:
fig,ax=plt.subplots()

ax.set_ylabel('Verdi bankinnskudd')
ax.set_xlabel('Tid T')

ax.plot(T,       account_balance(100,0.2,T),          label='Bankinnskudd, årlig forrentning')
_=ax.legend(loc='upper left')
```

La oss nå tenke oss at banken legger til renter n ganger i året. O, n=12, så legger banken til renter måndedlig. Vi kan da regne ut bankinnskuddet ved ulik antall forrentninger ber år:

```python
def account_balance_n(x,r,T,n):
    return x*(1+r/n)**(T*n)

ax.plot(T,       account_balance_n(100,0.2,T,4),          label='Kvartalsvis forrentning',     color='green')
ax.plot(T,       account_balance_n(100,0.2,T,12),         label='Måndedlig forrentning',       color='red')
ax.plot(T,       account_balance_n(100,0.2,T,52),         label='Ukentlig forrentning',        color='orange')
ax.plot(T,       account_balance_n(100,0.2,T,365),        label='Daglig forrentning',          color='purple')


print(f"""Differansen mellom årlig og daglig forrentning etter 20 år er {
    np.round(
    account_balance_n(100,0.2,10,365)-account_balance(100,0.2,10)
        ,2)
}""")

fig
```

I stedet for å regne ut med daglig forrentning kan vi bruke kontinuerlig, som vil si at vi bruker eksponentialfunksjonen

```python
def account_balance_exponential(x,r,T):
    return x*np.exp(r*T)

ax.plot(T,       account_balance_exponential(100,0.2,T),label='Kontinuerlig forrentet',      color='pink')

print(f"""Differansen mellom årlig og kontinuerlig forrentning etter 20 år er {
    np.round(
   account_balance_exponential(100,0.2,10)- account_balance_n(100,0.2,10,365)
        ,2)
}""")

fig
```
