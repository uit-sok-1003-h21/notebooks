# Funksjoner; linære, eksponential og logaritme

### Importerer nødvendige pakker

```python
import numpy as np
from matplotlib import pyplot as plt
np.arr
```

### Definerer funksjoner

Linær funksjon:

```python
def f(x):
    return 5+4*x
```

Eksponensiell funksjon:

```python
def g(x):
    return 0.01*np.exp(4*x)
```

Logaritmisk funksjon:

```python
def h(x):
    return np.log(0.01+4*x)
```

### Lager figur:

```python
x = np.linspace(0,5,100)

_=plt.plot(x,  f(x),                  label='f(x)')
_=plt.plot(x,  g(x),   color='green', label='g(x)')
_=plt.plot(x,  h(x),   color='red',   label='h(x)')

_=plt.legend(loc='upper center')
_=plt.ylim(0,20)
```

### Logaritme er den inverse funksjonen til eksponent, og motsatt:

Vi kan ta logaritmen til eksponentfunksjonen og eksponenten til logaritmefunksjonen, og få linære funksjoner 

```python
x = np.linspace(0,10,100)

_=plt.plot(x,          f(x),                      label='f(x)')
_=plt.plot(x,   np.log(g(x)),   color='green',label='g(x)')
_=plt.plot(x,   np.exp(h(x)),   color='red',  label='h(x)')

_=plt.legend(loc='upper center')
_=plt.ylim(0,50)
```
